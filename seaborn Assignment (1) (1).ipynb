### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

##### Scatter Plot:

A scatter plot is used to visualize the relationship between two variables to identify patterns, trends, or correlations.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
import matplotlib.pyplot as plt

# Create a scatter plot
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()
```

##### Bar Plot:

Bar plots are suitable for comparing categories or groups by showing the distribution of a single categorical variable.

```python
tips = sns.load_dataset('tips')
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill Amount by Day")
plt.show()
```

##### Histogram:

Histograms display the distribution of a single numerical variable, showing the frequency or count of values within specified bins.

```python
sns.histplot(data=iris, x="petal_length", kde=True)
plt.title("Distribution of Petal Length")
plt.show()
```

##### Box Plot (Box-and-Whisker Plot):

Box plots are used to visualize the distribution of a numerical variable and identify outliers and data spread.

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

##### Pair Plot:

Pair plots help visualize the pairwise relationships between multiple numerical variables in a dataset. It's useful for exploring data patterns and correlations

```python
sns.pairplot(iris, hue="species")
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Set up the subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Box plot for 'age' by 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')

# Box plot for 'fare' by 'pclass'
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataframe to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

# Set the title of the plot
plt.title('Passenger counts by month and year')

# Show the plot
plt.show()
```
